# SciPy: extra modules for scientific computation
https://www.scipy.org/

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Curve fitting - the most fundamental example: linear fit
```

```python
# A dataset....
x=np.linspace(-2,5,20)
def f(x,m,b):
    return m*x+b
y=f(x,2,1) + 0.9 * np.random.rand(len(x))
```

```python
# Ploting the dataset....
plt.figure(num=5)
plt.plot(x,y,'o')
plt.title('Linear y=mx+b data with noise')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Fitting the data....
popt, pcov = curve_fit(f, x, y)
perr = np.sqrt(np.diag(pcov))
y_fit = f(x, *popt)
plt.figure(num=6)
plt.plot(x, y, 'o', label='data')
plt.plot(x, y_fit, label='fit')
plt.annotate('m='+str('%.3f'%popt[0]+'$\pm$'+str('%.3f'%perr[0])), xy=(0.6, 0.35), xycoords='axes fraction')
plt.annotate('b='+str('%.3f'%popt[1]+'$\pm$'+str('%.3f'%perr[1])), xy=(0.6, 0.27), xycoords='axes fraction')
plt.legend(loc='best')
plt.show()
```

```python
#Now for something more complicated.......an exponential....
```

```python
def f(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
n = 60
x = np.linspace(0, 5, n)
y = f(x, 5, 2, 0.5) + 2 * np.random.rand(n)
```

```python
popt, pcov = curve_fit(f, x, y)
perr = np.sqrt(np.diag(pcov))
y_fit = f(x, *popt)
msd = np.sum((y - y_fit) ** 2) / n
```

```python
pnames = ['a', 'b', 'c']
results = ''
for name, value, error in zip(pnames, popt, perr):
    results += '{} = {:.2f}$\pm${:.2f}\n'.format(name, value, error)
results += 'MSD = {:.2f}'.format(msd)
plt.figure(num=7)
plt.plot(x, y, '.', label='data')
plt.plot(x, y_fit, label='fit: $ae^{-bx} + c$')
plt.annotate(results, xy=(0.7, 0.55), xycoords='axes fraction')
plt.legend()
plt.show()
```

```python
#Now let's analyse two synthetic signals to learn how to do time correlations
import math
import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
from scipy import signal
pi=math.pi

n=1000
sfreq=1e3 #sampling frequency
dt=1/sfreq  #sampling time
t=np.arange(0,n)*dt
f1=30.0  #in Hz
y1 = 2.0*np.sin(2.0*pi*f1*t) + 0.2*np.random.randn(n)
y2 = 2.0*np.sin(2.0*pi*f1*t+pi/4.0) + 0.2*np.random.randn(n)
plt.figure(num=8)
time_shift=(pi/4.0)/(2.0*pi*f1)
plt.plot(t,y1,label='y1')
plt.plot(t,y2,label='y2')
plt.title('correlated time series; time_shift='+str('%.5f'%time_shift))
plt.xlabel('Time [s]')
plt.ylabel('Amplitude [V]')
plt.legend(loc='best')
plt.xlim([0,0.1])
plt.show()
```

```python
# One clearly sees the time shift but if singals are noisy or with
# unknow frequency one cannot simply calculate the time shift by hand...
# Correlation comes to the rescue...careful get it normalised to [-1,1]
corr = signal.correlate(y1, y2,mode='same')/np.linalg.norm(y1)/np.linalg.norm(y2)
lags = signal.correlation_lags(len(y1), len(y2),mode='same')*dt
max_corr=np.argmax(np.abs(corr)) #highest in magnitude is what we want...
fig, ax = plt.subplots(2,num=9)
ax[0].plot(t,y1,label='y1')
ax[0].plot(t,y2,label='y2')
ax[0].set_xlim([0,0.1])
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Amplitude [V]')
ax[0].set_title('correlated time series; time_shift='+str('%.5f'%time_shift))
ax[0].legend(loc='best')
ax[1].plot(lags,corr)
ax[1].plot([lags[max_corr],lags[max_corr]],[-1,1],label='max_corr='+str('%.3f'%lags[max_corr]),color='red',linestyle='--')
ax[1].set_xlim([-0.05,0.05])
ax[1].set_xlabel('Time lag [s]')
ax[1].set_ylabel('Cross Corr. []')
ax[1].legend(loc='best')
fig.subplots_adjust(hspace=0.4)
plt.show()
```
